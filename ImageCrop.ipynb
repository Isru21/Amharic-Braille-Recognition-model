```python
from PIL import Image
import os

def crop_images(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through each file
    for file_name in files:
        # Check if the file is an image
        if file_name.lower().endswith('.jpg') or file_name.lower().endswith('.jpeg') or file_name.lower().endswith('.png'):
            file_path = os.path.join(folder_path, file_name)
            try:
                # Open the image
                image = Image.open(file_path)
                width, height = image.size
                
                # Calculate new width and height for maintaining aspect ratio
                target_width = 96
                target_height = int((target_width / width) * height)
                
                # Resize the image while maintaining aspect ratio
                resized_image = image.resize((target_width, target_height))
                
                # Calculate the cropping parameters
                left = 0
                top = (target_height - 170) // 2
                right = target_width
                bottom = top + 170
                
                # Crop the resized image to 96x170 pixels
                cropped_image = resized_image.crop((left, top, right, bottom))
                
                # Save the cropped image
                cropped_image.save(os.path.join(folder_path, f"{file_name.split('.')[0]}.jpg"), "JPEG")
                print(f"{file_name} cropped and saved successfully.")
                
            except Exception as e:
                print(f"Error processing {file_name}: {e}")

# Example usage
folder_path = "/Users/Hello/Downloads/newshit/brailcorrect"
crop_images(folder_path)
```

```python
from PIL import Image
import os

# Directory containing images
input_dir = "/Users/Hello/Downloads/newshit/brailcorrect"

# Directory to save black and white images
output_dir = "/Users/Hello/Downloads/newshit/brailcorrect"

# Threshold for converting to black and white
threshold = 128

# Check if output directory exists, if not, create it
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate over each file in the input directory
for file_name in os.listdir(input_dir):
    # Check if the file is an image (you can add more image formats as needed)
    if file_name.endswith(".jpg") or file_name.endswith(".png") or file_name.endswith(".jpeg"):
        # Open the image
        img_path = os.path.join(input_dir, file_name)
        img = Image.open(img_path)
        
        # Convert the image to grayscale
        grayscale_img = img.convert('L')
        
        # Apply threshold to convert to black and white
        bw_img = grayscale_img.point(lambda x: 0 if x < threshold else 255, '1')
        
        # Save the black and white image in the output directory
        output_path = os.path.join(output_dir, file_name)
        bw_img.save(output_path)
        print(f"{file_name} converted to black and white and saved.")
```
